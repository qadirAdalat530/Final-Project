```python
#Creating Position Group based on BP (Best Position)

player_list["BP"] = player_list["BP"].str.strip()

keeper = ["GK"]

defence = ["RB","RCB","CB","LCB","LB","RWB","LWB"]

midfield = ["RDM","CDM","LDM","RM","RCM","CM","LCM","LM","LAM","CAM","RAM"]

attack = ["RW","RF","CF","LF","LW","LS","ST","RS"]

def parse_position(position):
    if position in defence:
        return ("p_defence")
    elif position in midfield:
        return ("p_midfield")
    elif position in attack:
        return ("p_attack")
    else:
        return ("p_keeper")
    
player_list["Position_Gr"] = player_list["BP"].apply(lambda x:parse_position(x))
```

```python
#Creating Dummy Variables for Position_Gr
pl1 = pd.concat([player_list.drop('Position_Gr', axis=1), pd.get_dummies(player_list['Position_Gr'])], axis=1)
```

```python
#Creating Age Group based on Age

def parse_age(age):
    if age <= 20:
        return ("Group1")
    elif age >20 and age<=23:
        return ("Group2")
    elif age >23 and age<=27:
        return ("Group3")
    elif age >27 and age<=30:
        return ("Group4")
    elif age >30 and age<=35:
        return ("Group5")
    else:
        return ("Group6")
        
pl1["Age_Gr"] = pl1["Age"].apply(lambda x:parse_age(x))
```

```python
#Creating Dummy Variables for Age_Gr
pl2 = pd.concat([pl1.drop('Age_Gr', axis=1), pd.get_dummies(pl1['Age_Gr'])], axis=1);
```
